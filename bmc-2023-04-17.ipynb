*Blue Triangles*

Let us consider one white equilateral triangle (see Figure above, Step 0). Let us divide the sides of the triangle into three equal parts, and let us construct at
the bottom two similar blue equilateral triangles with a side that is a third of a side of the initial white triangle (see Figure, Step 1). If we repeat this procedure with each of the four white triangles obtained in Step 1, we will get 16 white triangles and 10 blue triangles in Step 2.

```python
import pandas as pd
```

```python
result = pd.DataFrame([[0, 1, 0, 1]], columns=["Step", "White", "Blue", "Total W"])

FB_previous = 0
FW_previous = 1
FT_previous = 1

for index in range(1, 7):
    FB_current = FW_previous * 2 + FB_previous
    FW_current = FW_previous * 4
    FT_current = FW_current + FT_previous
    FB_previous = FB_current
    FW_previous = FW_current
    FT_previous = FT_current

    result = pd.concat(
        [
            result,
            pd.DataFrame(
                [[index, FW_current, FB_current, FT_current]],
                columns=["Step", "White", "Blue", "Total W"],
            ),
        ],
        ignore_index=True,
    )

result.style.hide(axis="index")
```
